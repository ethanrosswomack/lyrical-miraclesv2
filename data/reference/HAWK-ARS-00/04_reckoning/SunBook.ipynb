```python
import nbformat as nbf
import os

# Define notebook name
notebook_name = "Sun_Tzu.ipynb"
base_dir = os.getcwd()

# Section headers based on known zip contents
sections = {
    [Albums=(HawkEye_Lyric_Book_Full/HawkEye_Lyric_Book_Full/01_Full_Disclosure_2020),
                (HawkEye_Lyric_Book_Full/HawkEye_Lyric_Book_Full/02_Behold_A_Pale_Horse_2020),
                (HawkEye_Lyric_Book_Full/HawkEye_Lyric_Book_Full/03_Milabs_2022),
                (HawkEye_Lyric_Book_Full/HawkEye_Lyric_Book_Full/04_Malicious_EP_2024),
                (HawkEye_Lyric_Book_Full/HawkEye_Lyric_Book_Full/05_Shadow_Banned_2024)],
    [Singles=(HawkEye_Lyric_Book_Full/HawkEye_Lyric_Book_Full/06_Singles)]

    
    'EverLights Rite' ==
    ["EverLights_Rite_Dissection"],
    
    [Dissection Archive=(EverLights_Rite_Dissection)],
    
    'Catalog Commentary' ==
    ["EverLights_Rite_Dissection"];
}

# Initialize notebook
nb = nbf.v4.new_notebook()
cells = []

# Intro
intro_text = (
    "# ðŸ“– The Book Of Sun Tzu\n"
    "Welcome to the master notebook â€” an organized compendium of **Hawk Eye The Rapper's** lyrical archive, commentary, and dissected meaning across every project.\n\n"
    "Use the links below to navigate into each project realm."
)
cells.append(nbf.v4.new_markdown_cell(intro_text))

# Build TOC with Markdown links
toc_md = "## ðŸ“š Table of Contents\n"
for label, path in sections.items():
    toc_md += f"### {label}\n"
    if os.path.isdir(path):
        for fname in sorted(os.listdir(path)):
            if fname.endswith(".ipynb"):
                rel_path = os.path.join(path, fname)
                toc_md += f"- [{fname}]({rel_path})\n"
    else:
        toc_md += f"- _(No notebooks found in `{path}`)_\n"
    toc_md += "\n"

cells.append(nbf.v4.new_markdown_cell(toc_md))

# Finalize notebook and save
nb['cells'] = cells
with open(notebook_name, 'w', encoding='utf-8') as f:
    nbf.write(nb, f)

print(f"âœ… Book_of_War.ipynb created successfully at {os.path.join(base_dir, notebook_name)}")
```
